```python
import threading
from src import *

def process_station(station_id):
    # Meteorolojik veriyi alma
    weather_data = get_weather_data(station_id)

    # Veriyi SQL'e kaydetme
    if weather_data:
        save_weather_data(weather_data)


WeatherObs.weather_by_istno(istNo)[0]

def process_stations(station_ids):
    # Her istasyon için ayrı thread oluşturma
    threads = []
    for station_id in station_ids:
        thread = threading.Thread(target=process_station, args=(station_id,))
        threads.append(thread)

    # Tüm thread'leri başlatma
    for thread in threads:
        thread.start()

    # Tüm thread'lerin bitmesini bekleme
    for thread in threads:
        thread.join()
```
